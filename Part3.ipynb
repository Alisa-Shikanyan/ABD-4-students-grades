Гипотеза 1: Учащиеся из частных школ имеют более высокие JAMB-баллы.


```python
private_scores = df[df['School_Type'] == 'Private']['JAMB_Score']
public_scores = df[df['School_Type'] == 'Public']['JAMB_Score']

from scipy import stats

t_stat, p_val = stats.ttest_ind(private_scores, public_scores, nan_policy='omit')
print(f"T-тест: t={t_stat:.2f}, p={p_val:.4f}")
```

p > 0,05 различия между группами статистически не значимы

Гипотеза 2: Уровень родительского образования влияет на JAMB-баллы студентов.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ANOVA тест
model = ols('JAMB_Score ~ C(Parent_Education_Level)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

Если PR(>F) < 0.05 — уровень образования родителей влияет на результаты JAMB.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Parent_Education_Level', y='JAMB_Score', data=df, order=[
    'None', 'Primary', 'Secondary', 'Tertiary'
])
plt.title("JAMB Баллы по уровню образования родителей")
plt.xticks(rotation=45)
plt.show()
```

Гипотеза 3: Пол студента влияет на JAMB-балл
Формулировка гипотез:
H₀ (нулевая гипотеза): Пол студента не влияет на его JAMB-балл (средние баллы равны).
H₁ (альтернативная гипотеза): Пол студента влияет на JAMB-балл (средние баллы различаются).


```python
# Разделение данных по полу
male_scores = df[df['Gender'] == 'Male']['JAMB_Score']
female_scores = df[df['Gender'] == 'Female']['JAMB_Score']

# Проверяем, есть ли данные в обеих группах
print(f"Количество мужчин: {len(male_scores)}")
print(f"Количество женщин: {len(female_scores)}")

# T-тест
from scipy import stats

t_stat, p_val = stats.ttest_ind(male_scores, female_scores, nan_policy='omit')

print(f"\nРезультаты t-теста:")
print(f"T-статистика: {t_stat:.2f}")
print(f"P-value: {p_val:.4f}")

# Интерпретация
alpha = 0.05
if p_val < alpha:
    print("✅ Отклоняем нулевую гипотезу: пол студента влияет на JAMB-балл.")
else:
    print("❌ Не можем отклонить нулевую гипотезу: нет значимой разницы между полами.")
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='JAMB_Score', data=df)
plt.title('JAMB Баллы по полу')
plt.xlabel('Пол')
plt.ylabel('JAMB Балл')
plt.grid(True)
plt.show()
```
